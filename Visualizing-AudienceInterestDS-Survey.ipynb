# Visualizing Audience Interest Survey

In this notebook we will visualize the results of a survey that was conducted to gauge an audience interest in different data science topics

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_survey = pd.read_csv('Topic_Survey_Assignment.csv',index_col=0)
df_survey.head()
```

```python
#sort dataframe
df_survey.sort_values(by='Very interested',ascending=False,inplace=True)

#convert it intro pecentage
df_survey = round((df_survey/2233)*100,2)
df_survey.head()
```

```python
ax = df_survey.plot(
    kind='bar',
    figsize=(20,10)
)

plt.title("Percentage of Respondent's Interest in DataScience Areas")
plt.legend(fontsize=14)

for p in ax.patches:
    height=p.get_height()
    ax.annotate(height,
                xy=(p.get_x()+p.get_width() / 2,height),
                ha = 'center',
                va='bottom',
                xytext=(0, 3),
                textcoords="offset points",
    )
```
